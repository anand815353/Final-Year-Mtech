#### TF computation

```python
def TF(word_dictionary, length):
    tf_dict = {}

    for word, count in word_dictionary.items():
        tf_dict[word] = count/length
    return tf_dict    
```

```python
def i_df(idf_dict,length):
    for word , val in idf_dict.items():
        idf_dict[word] = math.log10(length / float(val))
    return idf_dict
```

```python
def tf_idf(tf,idf_dict):
    tfidf = {}
    for word, val in tf.items():
        tfidf[word] = val*idf_dict[word]
    return tfidf
```
